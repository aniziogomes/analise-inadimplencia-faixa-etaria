```python
# Importando Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando Arquivo Excel
tabela = pd.read_excel(r"C:\Users\anizi\OneDrive\Documentos\Dados\bank_marketing.xlsx")

print(tabela.head()) # Mostrar as 5 primeiras linhas
```

```python
#informações Gerais da PLanilha
info = (tabela.info())

# Resumo Estatístico (média, mínimo, máximo)
missing_values = (tabela.describe())

# Verificando se há valores nulos em cada coluna
statistics = (tabela.isnull().sum())

info, missing_values, statistics
```

```python
# Extraindo a coluna irrelevante
tabela.limpeza = tabela.drop(columns=["Unnamed: 0"])

# verificando as mudanças
tabela.limpeza.head()
```

<b> PROCESSO ETL</b> Excluindo coluna irrelevante

```python
dataFrame = tabela

colunas_interesse = ["Idade", "Cliente_Devedor?"]
if all (coluna in dataFrame.columns for coluna in colunas_interesse):
    print("Colunas presentes")
else:
    raise ValueError("Faltam Colunas")
```

<b>Nota:</b> Extraindo apenas as colunas de interesse para análise ( Idade e Cliente Devedor )<br>

```python
dataFrame["Cliente_Devedor?"] = dataFrame["Cliente_Devedor?"].replace({"Sim": 1, "Não": 0})
#Transformando as colunas em númericas

print(dataFrame["Cliente_Devedor?"].unique())
# verificando alteração
```

```python
print("Estatísticas descritiva da idade:")
print(dataFrame["Idade"].describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(tabela["Idade"], kde=True, color="blue", bins=20)
plt.title("Distribuição de Idades", fontsize=16)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Frequência", fontsize=12)
plt.show()
```

<b>Público Majoritário de idade:</b>
A maior parte das pessoas está na faixa de 30 a 50 anos, com um pico próximo dos 30 anos, indica que é um público relativamente adulto.


```python
# Estilo dos gráficos
sns.set(style="whitegrid")
```

```python
devedores_proporcao = dataFrame["Cliente_Devedor?"].value_counts(normalize=True)
print("\nProporção de clietes devedores:")
print(devedores_proporcao)
```

<b>Explicação das Proporções:</b> Indica que 99,3% dos clientes não são inadimplentes, enquanto apenas 0,67% são inadimplentes. <br>
Significa que a cada 1.000 clientes, 6,71 são inadimplentes Indicando que a inadimplência nesse conjunto de dados é relativamente baixa.

```python
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ["0-20", "21-30", "31-40", "41-50", "51-60", "60+"]
dataFrame["Faixa_Etaria"] = pd.cut(dataFrame["Idade"], bins=bins, labels=labels, right=False)

proporcao_devedores = dataFrame.groupby("Faixa_Etaria")["Cliente_Devedor?"].mean()

plt.figure(figsize=(8, 5))
proporcao_devedores.plot(kind='bar', color='coral', edgecolor='black')
plt.title("Proporção de Clientes Devedores por Faixa Etária", fontsize=14)
plt.xlabel("Faixa Etária", fontsize=12)
plt.ylabel("Proporção de Devedores", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<b> NOTA: </b> A faixa etária 51-60 anos apresenta a maior proporção de clientes devedores, seguida pelas faixas 21-30 anos e 41-50 anos.
Isso sugere que adultos jovens e de meia-idade têm maior probabilidade de inadimplência.
